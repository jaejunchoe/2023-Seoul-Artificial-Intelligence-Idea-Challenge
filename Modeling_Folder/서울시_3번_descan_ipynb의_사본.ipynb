```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np
from sklearn.preprocessing import StandardScaler


# 필요한 열만 선택
X = df_test1[['독거노인 소계', '주거 현원']]

# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 가능한 eps와 min_samples 값들
eps_values = np.linspace(0.1, 1.0, 10)
min_samples_values = range(2, 11)

best_silhouette_score = -1
best_eps = None
best_min_samples = None

# 최적의 조합 찾기
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        y_pred = dbscan.fit_predict(X_scaled)

        # 클러스터가 1개 이상 존재해야 실루엣 스코어 계산 가능
        if len(np.unique(y_pred)) > 1:
            silhouette_avg = silhouette_score(X_scaled, y_pred)
            if silhouette_avg > best_silhouette_score:
                best_silhouette_score = silhouette_avg
                best_eps = eps
                best_min_samples = min_samples

print('최적 eps:', best_eps)
print('최적 min_samples:', best_min_samples)
print('최적 실루엣 스코어:', best_silhouette_score)
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# eps와 min_samples의 값을 변경하여 시도
for eps in [0.1, 1.0, 10]:
    for min_samples in [2, 3, 4]:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(X_scaled)
        df_test1['Cluster'] = clusters

        # 클러스터 개수 (노이즈 포인트 제외)
        n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
        print(f'eps: {eps}, min_samples: {min_samples}, number of clusters: {n_clusters}')

        # 결과 시각화
        plt.figure(figsize=(10, 6))
        plt.scatter(df_test1['독거노인 소계'], df_test1['주거 현원'], c=df_test1['Cluster'], cmap='viridis', s=50)
        plt.colorbar().set_label('Cluster')
        plt.xlabel('독거노인 소계')
        plt.ylabel('주거 현원')
        plt.title(f'DBSCAN Clustering (eps: {eps}, min_samples: {min_samples})')
        plt.grid(True)
        plt.show()
```

```python
## 비례, 반비례 관계 확인
```

```python
import pandas as pd
from scipy.stats import pearsonr

# 클러스터별로 데이터를 그룹화
grouped = df_filtered.groupby('Cluster')

# 각 클러스터에서의 상관 관계를 계산
for name, group in grouped:
    correlation_coefficient, _ = pearsonr(group['독거노인 소계'], group['주거 현원'])
    if correlation_coefficient > 0:
        relationship = '비례'
    elif correlation_coefficient < 0:
        relationship = '반비례'
    else:
        relationship = '관계 없음'

    print(f'클러스터 {name}: {relationship} 관계, 상관 계수: {correlation_coefficient:.2f}')
```

```python
# 클러스터별로 데이터를 그룹화
cluster_groups = df_filtered.groupby('Cluster')

# 각 클러스터별로 동별(2) 열의 값들을 출력
for cluster, group in cluster_groups:
    print(f'Cluster {cluster}:')
    print(group['동별(2)'].tolist())
    print()  # 빈 줄을 출력하여 클러스터별 출력을 구분
```
